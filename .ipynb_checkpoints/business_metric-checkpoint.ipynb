Faria sentido analisar simplesmente qual range de frete teve mais compras? entendo que  o resultado seria enviesado ja que poucos produtos serão ofericidos por um frete muito alto ou muito baixo, por ex.. 

Para fazer uma comparação justa vou tentar fazer uma comparação das vendas do mesmo produto com diferentes fretes. 

```python

# Transformando os clusters de preço em coluna para serem somados quando agruparmos pelo produto
raw_freight_df_3 = pd.concat([raw_freight_df_2,pd.get_dummies(raw_freight_df_2.freight_cluster)],axis = 1)

# Agrupando por produto
raw_freight_gp = raw_freight_df_3.groupby('product_id',as_index = False).sum()
raw_freight_gp.drop('freight_value',axis = 1,inplace = True) # dropando frete somado

# Total de pedidos
raw_freight_gp['qnt_pedidos'] = raw_freight_df_3.groupby('product_id').price.count().values

# Usando apenas produtos com mais de 10 pedidos na análise
raw_freight_gp = raw_freight_gp[raw_freight_gp['qnt_pedidos'] >10]


raw_freight_gp.sample(5)
```

```python
# Transformando a somatoria de pedidos em porcentagem de pedidos 
# i.e qnts % das compras do produto x foram realizadas nessa faixa de frete? 

for col in raw_freight_gp.columns[2:-1]:
    raw_freight_gp[col] = raw_freight_gp.apply(lambda x: 100*x[col]/x['qnt_pedidos'],axis = 1)
  
raw_freight_gp.sample(3)
```
